```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
import json
from datetime import date

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "../output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#API Call to get weather conditions for the cities

#Start with base URL
baseUrl='https://api.openweathermap.org/data/2.5/weather?q='
weatherKey='&appid='+ weather_api_key  #API_Key
units='&units=imperial'                #Imperial Units

#Creating the arrays to append to late
foundCities=[]
citiesCountry=[]
dates=[]
citiesLat=[]
citiesLng=[]
citiesMaxTemp=[]
citiesHumidity=[]
citiesCloudiness=[]
citiesWind=[]
counter=1
setCounter=1


# -----Teste
# city='Lisbon'
# queryUrl=baseUrl+city+units+weatherKey
# r=requests.get(queryUrl).json()
# print(json.dumps(r,indent=4))

# foundCities.append(r['name'])
# citiesCountry.append(r['sys']['country'])
# dates.append(r['dt'])
# citiesLat.append(r['coord']['lat'])
# citiesLng.append(r['coord']['lon'])
# citiesMaxTemp.append(r['main']['temp'])
# citiesHumidity.append(r['main']['humidity'])
# citiesCloudiness.append(r['clouds']['all'])
# citiesWind.append(r['wind']['speed'])
# ------Teste

#Printing the first line of the output
print('Beginning Data Retrieval\n-------------------------')

#For loop to go over every single city of the cities array
for index,city in enumerate(cities):
    #When counter is = 50, stop for 90 secs (to not reach the API limit of call), increase setCounter and reset counter
    if counter == 50:
        time.sleep(90)
        setCounter+=1
        counter=1
    #If counter is <> 50, do the API call
    else:
        queryUrl=baseUrl+city+units+weatherKey
        r=requests.get(queryUrl).json()
        
        #THandling errors with the try and appending the data to the arrays created before
        try:
            foundCities.append(r['name'])
            citiesCountry.append(r['sys']['country'])
            dates.append(r['dt'])
            citiesLat.append(r['coord']['lat'])
            citiesLng.append(r['coord']['lon'])
            citiesMaxTemp.append(r['main']['temp'])
            citiesHumidity.append(r['main']['humidity'])
            citiesCloudiness.append(r['clouds']['all'])
            citiesWind.append(r['wind']['speed'])
            #printing the city name, city number and the set call they were part of
            print(f'City number {counter} of set {setCounter} | {city}')
            counter+=1
        #If error print city not found
        except:
            print(f'City was not found') 
print('-------------------------\nFinished Data Retrieval\n-------------------------')
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#Create DataFrame with the data collected in the arrays
citiesWeather=pd.DataFrame({
    'City':foundCities,
    'Country':citiesCountry,
    'Date':dates,
    'Lat':citiesLat,
    'Lon':citiesLng,
    'Max Temp (F)':citiesMaxTemp,
    'Humidity (%)':citiesHumidity,
    'Cloudiness (%)':citiesCloudiness,
    'Wind Speed (mph)':citiesWind
})

#count the values for every column 
citiesWeather.count()
```

## Linear Regression

```python
# Create Northern and Southern Hemisphere DataFrames
NorthCities=citiesWeather.loc[citiesWeather['Lat']>=0,:]
SouthCities=citiesWeather.loc[citiesWeather['Lat']<0,:]
```

```python
#Creating a function to create Linear Regression plots. Only inputs are the Hemisphere, the y-values and the y-label.

def LinReg (Hemisphere,y,y_label):
    # defining x and y values
    x_axis=Hemisphere['Lat']
    y_axis=Hemisphere[y]
    
    #Calculating the values for the lenear regression
    (slope, interception, r, p, std_err)=linregress(x_axis,y_axis)
    
    #Creating the values for y based on the xvalues to create the regression line to map it
    line=slope*x_axis+interception
    
    #writing the equation of the regression line
    line_eq=f'y={round(slope,2)}x+{round(interception)}'
    
    #Printing the r2 value
    print(f'The r-squared is: {r}')
    
    #creating the scatter plot
    plt.scatter(x_axis,y_axis)
    plt.plot(x_axis,line,'r-')
    plt.xlabel('Latitude')
    plt.ylabel(y_label)
    plt.title(f'Latitude vs {y_label}    ({current})')
    
    # splited north and south hemispher just to place the regression line in the plot since I could not find a good place
    # to put it for both hemispheres
    if max(x_axis) > 0:
        plt.annotate(line_eq,(0,(interception - (interception/2))),fontsize=13,color='red',bbox=dict(boxstyle="round", fc="w"))
        plt.show()
    else:
        yAnno=(slope*min(x_axis)+interception)+((max(y_axis)-min(y_axis))/5)
        plt.annotate(line_eq,(-56,yAnno),fontsize=13,color='red',bbox=dict(boxstyle="round", fc="w"))
        plt.show()
    
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
#Using the created function to create the plot and the regression line

LinReg(NorthCities,'Max Temp (F)','Max Temperature (F)')
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
#Using the created function to create the plot and the regression line

LinReg(SouthCities,'Max Temp (F)','Max Temperature (F)')
```

```python
#ANLYSIS

#In the above 2 plots I'm analyzing the impct of the latitude on the maximum temperature for both the Northern and Southern
#Hemispheres.
#The Northen plot shows a r2 value of almost -0.9 which means that based on the latitude we can accurately predict the max 
#temperature, at least at this time of the year. The negative just means that it's a negative correlation (negative slope)
#The Southern graphic has a significantly lower r2, meaning that it isn't as predictable (based on the latitude alone) as
#the cities from the North. It has a possitive correlation meaning the higher the latitude, the higher the temperature.
```
